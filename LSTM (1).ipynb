```python
import numpy as np
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)
```

```python
# Example sequence: 1, 2, 3
sequence = [1, 2, 3]
```

```python
#dimensions
input_dim = 1
hidden_dim = 1

# Weights
Wf = 0.5
Wi = 0.6
Wo = 0.7
Wc = 0.8

Uf = 0.1
Ui = 0.1
Uo = 0.1
Uc = 0.1

# Biases
bf = 0
bi = 0
bo = 0
bc = 0
```

```python
# Initial hidden state and cell state
h = 0
c = 0
```

```python
for t, x in enumerate(sequence):

    print("Time Step:", t+1)

    # Forget Gate
    f = sigmoid(Wf*x + Uf*h + bf)

    # Input Gate
    i = sigmoid(Wi*x + Ui*h + bi)

    # Candidate Cell State
    c_tilde = tanh(Wc*x + Uc*h + bc)

    # Cell State Update
    c = f*c + i*c_tilde

    # Output Gate
    o = sigmoid(Wo*x + Uo*h + bo)

    # Hidden State
    h = o * tanh(c)

    print("Forget gate:", f)
    print("Input gate:", i)
    print("Candidate state:", c_tilde)
    print("Cell state:", c)
    print("Output gate:", o)
    print("Hidden state:", h)
    print("----------------------")
```

```python
# Linear Transformation
Wy = 1.5
by = 0.5

y_pred = Wy * h + by

print("Predicted Next Value:", y_pred)
```
